```python
# Importing Raw Data

df = pd.read_csv("train.csv")

df.isnull().sum()
```

```python
# Number of missing values for Mileage column

print(df['Mileage'].isna().value_counts())

# Removing those 2 rows as this does not effect on raw data

df = df[df['Mileage'].notna()]
```

```python
# Number of missing values for Engine
print(df['Engine'].isna().value_counts())

```

```python
#Removing 36 rows as this does not effect on raw data
df = df[df['Engine'].notna()]
```

```python
# Checking number of missing values for Power
print(df['Power'].isna().value_counts())
```

```python
#Removing 36 rows as this does not effect on raw data

df = df[df['Power'].notna()]
```

```python
# Checking number of missing values for Seats

print(df['Seats'].isna().value_counts())
```

```python
#Removing those 2 rows as this does not effect on raw data

df = df[df['Seats'].notna()]
```

```python
#Checking number of missing values for New_Price

print(df['New_Price'].isnull().value_counts())
```

```python
# As the values in New_Price column has more variance so we can't perform data analysis if we drop this column. So, Replacing the NA Values with the median rather than mean is appropriate. 

df['New_Price'] = df['New_Price'].str.replace(" Lakh", "").str.replace(" Cr","").astype(float)

New_price_median = df[df['New_Price'].notna()]['New_Price'].median()

print(New_price_median)
```

```python
#So now we are Replacing NA values with median in New_Price column

df['New_Price'] = df['New_Price'].fillna(value = New_price_median)

df.head()
```

```python
#Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).


df['Mileage'] = df['Mileage'].str.replace(" kmpl","").str.replace(" km/kg","").astype(float)

df['Engine'] = df['Engine'].str.replace(" CC","").astype(int)

df['Power'] = df['Power'].str.replace(" bhp","").astype(float)

df.head()
```

```python

# Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value


df['Fuel_Type'].value_counts()

df['Transmission'].value_counts()

# one hot encoding for Fuel_Type and Transmission

df = pd.get_dummies(
    data=df,
    columns=["Fuel_Type", "Transmission"],
    prefix=["Fuel_Type", "Transmission"])

df.head()
```

```python
# Creating one more feature and adding to the original data frame i.e., current age of car calculating current age of car

df['Current_Age'] = 2023 - df['Year']

df.head()
```
